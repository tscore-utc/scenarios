```python
import pandas as pd
import numpy as np
#from bs4 import BeautifulSoup #to beautify xml output
#from lxml import etree as ET #read and write xml files
pd.options.display.max_columns=50

import os
from os import listdir
from os.path import isfile, join

import time

import geopandas as gp
```

```python
taz_lbnl=gp.read_file('../LBNL_Files/sf-light-tazs.shp')
taz_lbnl=taz_lbnl.to_crs('EPSG:26910')
taz_lbnl=taz_lbnl.rename(columns={'taz':'LBNL_TAZ'})
```

```python
taz_sfchamp=gp.read_file('../SF-CHAMP Outputs/TAZ2454_clean.shp')
taz_sfchamp=taz_sfchamp.to_crs('EPSG:26910')
taz_sfchamp=taz_sfchamp.rename(columns={'TAZ':'SFCHAMP_TAZ'})
taz_sfchamp['SFCHAMP_TAZ_AREA']=taz_sfchamp['geometry'].area
```

```python
overlaid_taz=gp.overlay(taz_sfchamp,taz_lbnl,how='union',keep_geom_type=False)
overlaid_taz['OVERLAY_AREA']=overlaid_taz['geometry'].area
overlaid_taz['OVERLAY_RATIO']=overlaid_taz['OVERLAY_AREA']/overlaid_taz['SFCHAMP_TAZ_AREA']

overlaid_taz=overlaid_taz[['SFCHAMP_TAZ','LBNL_TAZ','OVERLAY_RATIO','SFCHAMP_TAZ_AREA','OVERLAY_AREA']]

overlaid_taz.head()
```

```python
lbnl_parking=pd.read_csv('../SF_CHAMP_Converted/sf-taz-parking-nochargers-updated.csv')
lbnl_parking=lbnl_parking.rename(columns={'numStalls':'orig_numStalls'})

lbnl_parking.head()
```

```python
overlaid_taz_parking=pd.merge(overlaid_taz,lbnl_parking,
                              left_on=['LBNL_TAZ'],right_on=['taz'],how='left')

overlaid_taz_parking['numStalls']=round(overlaid_taz_parking['OVERLAY_RATIO']*overlaid_taz_parking['orig_numStalls'])

overlaid_taz_parking=overlaid_taz_parking[overlaid_taz_parking['SFCHAMP_TAZ'].notnull()]

overlaid_taz_parking.drop(columns=['taz'],inplace=True)

overlaid_taz_parking.tail()
```

```python
overlaid_taz_parking['SFCHAMP_TAZ']=overlaid_taz_parking['SFCHAMP_TAZ'].astype(str)
overlaid_taz_parking=overlaid_taz_parking.rename(columns={'SFCHAMP_TAZ':'taz'})
overlaid_taz_parking['taz']=overlaid_taz_parking['taz'].apply(lambda x:x.split('.')[0])


overlaid_taz_parking['parkingType']=overlaid_taz_parking['parkingType'].fillna('Public')
overlaid_taz_parking['pricingModel']=overlaid_taz_parking['pricingModel'].fillna('Block')
overlaid_taz_parking['chargingPointType']=overlaid_taz_parking['chargingPointType'].fillna('NoCharger')
overlaid_taz_parking['feeInCents']=overlaid_taz_parking['feeInCents'].fillna(0)
overlaid_taz_parking['reservedFor']=overlaid_taz_parking['reservedFor'].fillna('Any')
overlaid_taz_parking['numStalls']=overlaid_taz_parking['numStalls'].fillna(1000000)

overlaid_taz_parking_df=overlaid_taz_parking[['taz','parkingType','pricingModel','chargingPointType','numStalls','feeInCents',
                     'reservedFor']]
overlaid_taz_parking_df.head()
```

```python
overlaid_taz_parking_grouped=overlaid_taz_parking_df.groupby(['taz','parkingType','pricingModel','chargingPointType','reservedFor']).agg({'numStalls':'sum',
                                                                                                                                               'feeInCents':'mean'}).reset_index()

overlaid_taz_parking_grouped.to_csv('../SF_CHAMP_Converted/sfchamp_parking_updated.csv',index=False)
```

```python
taz_parking=pd.read_csv('../SF-CHAMP outputs/tazdata2015.csv')
taz_parking.head()
```

```python
#metadata source https://www.sfcta.org/sites/default/files/content/Planning/VanNess_BRT_EIR/FEIR-FEIS/Tech%20Memos/Transportation-Traffic/VN%20BRT%20Vehicular%20Tech%20Memo%20Appendices_CHS_7_3_2013_Final_LU.pdf
taz_parking=taz_parking[['SFTAZ',
 'MTCTAZ',
 'FREEPARK', #share of free parking
 'SUBPARK', #share of subsidized parking
 'PAIDPARK', #share of paid parking
 'PPAYING', #percent paying for parking
 'PRKAVIND', #parking availability index
 'ONSTREET', #fraction of on street parking
 'OFSTREET', #fraction of off street parking
 'PRKCSTWH', #hourly parking cost for work tours
 'PRKCSTOH', #hourly parking cost for other tours
 'TOTPARK']] #total number of parking
taz_parking.head()
```

```python
taz_parking['Workplace']=np.ceil(taz_parking['TOTPARK']*taz_parking['SUBPARK'])
taz_parking['Public']=np.ceil(taz_parking['TOTPARK']*taz_parking['PAIDPARK'])
taz_parking['Residential']=np.ceil(taz_parking['TOTPARK']*taz_parking['FREEPARK'])

taz_parking['Worktour_Cost']=taz_parking['PRKCSTWH']*100
taz_parking['Other_Cost']=taz_parking['PRKCSTOH']*100
# taz_parking['Worktour_Cost']=np.where(taz_parking['PRKCSTWH']==0,
#                                      100,
#                                      taz_parking['PRKCSTWH']*100)

# taz_parking['Other_Cost']=np.where(taz_parking['PRKCSTOH']==0,
#                                      100,
#                                      taz_parking['PRKCSTOH']*100)

taz_parking['Free_Cost']=0
taz_parking.head()
```

```python
numstall=pd.melt(taz_parking,
                 id_vars=['SFTAZ'],
                 value_vars=['Workplace','Public','Residential'],
                var_name='parkingType',
                value_name='numStalls')
numstall=numstall.sort_values('SFTAZ')
numstall.head()
```

```python
numstall['reservedFor']='any'
numstall['chargingType']='NoCharger'
numstall.head()
```

```python
parking_fees=pd.melt(taz_parking,
                 id_vars=['SFTAZ'],
                 value_vars=['Worktour_Cost','Other_Cost','Free_Cost'],
                var_name='parkingType',
                value_name='feeInCents')
parking_fees=parking_fees.sort_values('SFTAZ')
parking_fees.head()
```

```python
def parking_type(x):
    if x=='Worktour_Cost':
        return "Workplace"
    elif x=='Free_Cost':
        return "Residential"
    else:
        return "Public"
    
parking_fees['parkingType']=parking_fees['parkingType'].apply(parking_type)
parking_fees.head()
```

```python
taz_parking_constrained=pd.merge(numstall, parking_fees,
                                on=['SFTAZ','parkingType'],how='outer')

taz_parking_constrained['pricingModel']='FlatFee'
taz_parking_constrained=taz_parking_constrained.rename(columns={'SFTAZ':'taz'})
taz_parking_constrained.head()
```

```python
taz_parking_constrained.to_csv('../SF_CHAMP_Converted/taz_parking_constrained.csv.gz',
                               index=False,
                              compression='gzip')
```

```python
taz_parking_unconstrained=taz_parking_constrained
taz_parking_unconstrained['numStalls']=2e6
taz_parking_unconstrained['feeInCents']=0
#public unsubsizided parking for 2 hour
#workplace 8 hours of parking cost for work tours hourly*8*discounted by the subsidized and free (average)
#residential free
taz_parking_unconstrained.head()
```

```python
taz_parking_unconstrained.to_csv('../SF_CHAMP_Converted/taz_parking_unconstrained.csv.gz',
                               index=False,
                              compression='gzip')
```

```python
ridehail=gp.read_file('../SF_CHAMP_Converted/DriverFleet_5min_roundoff_TAZCentroid_v1.0.geojson')

ridehail.head()
```

```python
ridehail['initialLocationX']=ridehail.geometry.centroid.x
ridehail['initialLocationY']=ridehail.geometry.centroid.y
ridehail.head()
```

```python
ridehail['shift_start_time']=ridehail['start_shift'].apply(lambda x: x.split(' ')[2])
ridehail['shift_end_time']=ridehail['end_shift'].apply(lambda x: x.split(' ')[2])


ridehail.head()
```

```python
import datetime as dt
```

```python
def seconds_after_midnight(x):
    hh=int(x.split(':')[0])
    mm=int(x.split(':')[1])
    ss=int(x.split(':')[2])
    
    return str(hh*3600+mm*60+ss)

ridehail['shift_start_seconds_after_midnight']=ridehail['shift_start_time'].apply(seconds_after_midnight)

ridehail['shift_end_seconds_after_midnight']=ridehail['shift_end_time'].apply(seconds_after_midnight)

ridehail.head()
```

```python
ridehail['shifts']='{'+ridehail['shift_start_seconds_after_midnight']+':'+ridehail['shift_end_seconds_after_midnight']+'}'

ridehail.head()
```

```python
ridehail['rideHailManagerId']='GlobalRHM'
ridehail['geofenceX']=''
ridehail['geofenceY']=''
ridehail['geofenceRadius']=''
ridehail['vehicleType']='Car'
```

```python
ridehail=ridehail[['id','rideHailManagerId','vehicleType','initialLocationX',
         'initialLocationY','shifts','geofenceX','geofenceY','geofenceRadius']]
ridehail.head()
```

```python
ridehail['index']=ridehail.index
ridehail['index']=ridehail['index'].astype(str)
ridehail['new_id']=ridehail['index'].apply(lambda x:'rideHailVehicle-'+x )   
ridehail.head()
```

```python
ridehail[['new_id','rideHailManagerId','vehicleType','initialLocationX',
         'initialLocationY','shifts','geofenceX','geofenceY','geofenceRadius']].to_csv('../SF_CHAMP_Converted/ridehailFleetUpdated.csv',
                                                                                      index=False)
```
