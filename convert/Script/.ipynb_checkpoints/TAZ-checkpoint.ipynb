```python
taz_parking=pd.read_csv('../SF-CHAMP outputs/tazdata2015.csv')
taz_parking.head()
```

```python
#metadata source https://www.sfcta.org/sites/default/files/content/Planning/VanNess_BRT_EIR/FEIR-FEIS/Tech%20Memos/Transportation-Traffic/VN%20BRT%20Vehicular%20Tech%20Memo%20Appendices_CHS_7_3_2013_Final_LU.pdf
taz_parking=taz_parking[['SFTAZ',
 'MTCTAZ',
 'FREEPARK', #share of free parking
 'SUBPARK', #share of subsidized parking
 'PAIDPARK', #share of paid parking
 'PPAYING', #percent paying for parking
 'PRKAVIND', #parking availability index
 'ONSTREET', #fraction of on street parking
 'OFSTREET', #fraction of off street parking
 'PRKCSTWH', #hourly parking cost for work tours
 'PRKCSTOH', #hourly parking cost for other tours
 'TOTPARK']] #total number of parking
taz_parking.head()
```

```python
taz_parking['Workplace']=np.ceil(taz_parking['TOTPARK']*taz_parking['SUBPARK'])
taz_parking['Public']=np.ceil(taz_parking['TOTPARK']*taz_parking['PAIDPARK'])
taz_parking['Residential']=np.ceil(taz_parking['TOTPARK']*taz_parking['FREEPARK'])

taz_parking['Worktour_Cost']=taz_parking['PRKCSTWH']*100
taz_parking['Other_Cost']=taz_parking['PRKCSTOH']*100
# taz_parking['Worktour_Cost']=np.where(taz_parking['PRKCSTWH']==0,
#                                      100,
#                                      taz_parking['PRKCSTWH']*100)

# taz_parking['Other_Cost']=np.where(taz_parking['PRKCSTOH']==0,
#                                      100,
#                                      taz_parking['PRKCSTOH']*100)

taz_parking['Free_Cost']=0
taz_parking.head()
```

```python
numstall=pd.melt(taz_parking,
                 id_vars=['SFTAZ'],
                 value_vars=['Workplace','Public','Residential'],
                var_name='parkingType',
                value_name='numStalls')
numstall=numstall.sort_values('SFTAZ')
numstall.head()
```

```python
numstall['reservedFor']='any'
numstall['chargingType']='NoCharger'
numstall.head()
```

```python
parking_fees=pd.melt(taz_parking,
                 id_vars=['SFTAZ'],
                 value_vars=['Worktour_Cost','Other_Cost','Free_Cost'],
                var_name='parkingType',
                value_name='feeInCents')
parking_fees=parking_fees.sort_values('SFTAZ')
parking_fees.head()
```

```python
def parking_type(x):
    if x=='Worktour_Cost':
        return "Workplace"
    elif x=='Free_Cost':
        return "Residential"
    else:
        return "Public"
    
parking_fees['parkingType']=parking_fees['parkingType'].apply(parking_type)
parking_fees.head()
```

```python
taz_parking_constrained=pd.merge(numstall, parking_fees,
                                on=['SFTAZ','parkingType'],how='outer')

taz_parking_constrained['pricingModel']='FlatFee'
taz_parking_constrained=taz_parking_constrained.rename(columns={'SFTAZ':'taz'})
taz_parking_constrained.head()
```

```python
taz_parking_constrained.to_csv('../SF_CHAMP_Converted/taz_parking_constrained.csv.gz',
                               index=False,
                              compression='gzip')
```

```python
taz_parking_unconstrained=taz_parking_constrained
taz_parking_unconstrained['numStalls']=2e6
taz_parking_unconstrained['feeInCents']=0
#public unsubsizided parking for 2 hour
#workplace 8 hours of parking cost for work tours hourly*8*discounted by the subsidized and free (average)
#residential free
taz_parking_unconstrained.head()
```

```python
taz_parking_unconstrained.to_csv('../SF_CHAMP_Converted/taz_parking_unconstrained.csv.gz',
                               index=False,
                              compression='gzip')
```
